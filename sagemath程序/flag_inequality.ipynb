```python
# xor 10 bit
Points = []
for i in range(1024):
    bin_str = (10-len(bin(i)[2:]))*'0'+bin(i)[2:]
    point = []
    tmp_sum = 0
    for j in range(10):
        point.append(int(bin_str[j]))
        tmp_sum += int(bin_str[j])
    tmp_sum = tmp_sum % 2
    point.append(tmp_sum)
    Points.append(point)
# print(len(Points), Points)

Poly1 = Polyhedron(vertices=Points)
ineq_list1 = []
for ineq in Poly1.inequality_generator ():
#     print(ineq, list(ineq))
    ineq_list1.append(list(ineq))
# print(ineq_list1, len(ineq_list1))

print("Points:")
for point in Points:
    print(point)
    
print("ineqs:")
for ineq in ineq_list1:
    print(ineq)
```
