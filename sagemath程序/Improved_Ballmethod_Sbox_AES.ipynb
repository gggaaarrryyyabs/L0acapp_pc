```python
# 按照文章的参数写的
from sage.crypto.sboxes import AES
d_0 = 1
r = 2
n = 16
M_i = 50
M_prime = 800
M = 1610
S = set([256*diff+AES(x)^^AES(x^^diff) for diff in range(256) for x in range(256)])
F2n = set([256*x+y for x in range(256) for y in range(256)])

S_complement = F2n.difference(S)
nchoose2 = unordered_tuples([0..n-1], 2)
B_x0r2 = [0]
for i in [0..len(nchoose2)-1]:
    if nchoose2[i][0] != nchoose2[i][1]:
        B_x0r2.append(2**nchoose2[i][0]+2**nchoose2[i][1])
    else:
        B_x0r2.append(2**nchoose2[i][0])
# print(len(S_complement))
```

```python
# 5h左右的运行时间吧
flag = 0
inequation_list=[]
for y in S_complement:
    Byr = Set([x for x in B_x0r2]).intersection(Set([y^^s for s in S_complement]))
#     print(len(Byr))
    Pyr_list = []
    Byr_complement = F2n.difference(Byr)
    for z in Byr:
        for i in range(16):
            if z^^(2**i) in Byr_complement:
                Pyr_list.append(z^^(2**i))
    Pyr = set(Pyr_list)
    p = MixedIntegerLinearProgram(maximization=True, solver = "GLPK")
    ba_idx = p.new_variable(integer=True, nonnegative=False)
    # the range of b 
    p.set_max(ba_idx[0],M_prime)
    p.set_min(ba_idx[0],-M_prime)
    
    # the range of a_i
    for i in [1..n]:
        p.set_max(ba_idx[i],M_i)
        p.set_min(ba_idx[i],-M_i)
    
    # the range of dx
    for j in [n+1..n+len(Byr)]:
        p.set_max(ba_idx[j],1)
        p.set_min(ba_idx[j],0)
    p.set_min(ba_idx[n+1],1)
    
    for x in Pyr:
        x_binary = x.bits()
        if len(x_binary) != n:
            [x_binary.append(0) for i in range(n-len(x_binary))]
        p.add_constraint(-1*ba_idx[0] + sum([x_binary[index-1]*ba_idx[index] for index in [1..n]]) >= 0)
        
    j = 0
    for x in Byr:
        x_binary = x.bits()
        if len(x_binary) != n:
            [x_binary.append(0) for i in range(n-len(x_binary))]
        p.add_constraint(-1*ba_idx[0] + sum([x_binary[index-1]*ba_idx[index] for index in [1..n]]) - M*(1-ba_idx[n+1+j]) <= -1)
        j = j+1
    p.set_objective(sum([ba_idx[i] for i in [n+1..n+len(Byr)]]))
#     print("size of Byr",len(Byr))
    print(p.solve())
    inequation = p.get_values(ba_idx)
    seq = [inequation[i] for i in range(n+1)]
#     print(seq)
    if seq in inequation_list:
        print("repeated:",seq)
    else:
        inequation_list.append(seq)
    print("now the number of inequalities is ",len(inequation_list))
#     print(p.solve())
    if len(inequation_list) >= 0:
        result = str(inequation_list)
        file = open('sagetest2.txt','w')
        file.write(result)
        file.close()
```
